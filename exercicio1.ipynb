```python
from collections import deque

def estado_inicial():
    return (1, frozenset())

def estado_objetivo(salas_aspiradas):
    return len(salas_aspiradas) == 3

def acoes(estado):
    posicao_atual, salas_aspiradas = estado
    proximas_acoes = []

    proximas_acoes.append(('aspirar', posicao_atual))

    adjacencias = {
        1: [2, 3],
        2: [1, 3],
        3: [1, 2]
    }
    
    for proxima_sala in adjacencias[posicao_atual]:
        proximas_acoes.append(('mover', proxima_sala))
    
    return proximas_acoes

def aplicar_acao(estado, acao):
    posicao_atual, salas_aspiradas = estado
    tipo_acao, destino = acao

    if tipo_acao == 'aspirar':
        novas_salas_aspiradas = salas_aspiradas.union({destino})
        return (posicao_atual, frozenset(novas_salas_aspiradas))
    
    elif tipo_acao == 'mover':
        return (destino, salas_aspiradas)

def busca_em_largura():
    inicio = estado_inicial()
    fila = deque([inicio])
    visitados = set()
    visitados.add(inicio)

    while fila:
        estado_atual = fila.popleft()
        if estado_objetivo(estado_atual[1]):
            return estado_atual

        for acao in acoes(estado_atual):
            novo_estado = aplicar_acao(estado_atual, acao)
            if novo_estado not in visitados:
                visitados.add(novo_estado)
                fila.append(novo_estado)

    return None

solucao = busca_em_largura()
print(f"Estado final: {solucao}")
```
