```python
from collections import deque

def estado_inicial():
    return (frozenset(['fox', 'chicken', 'corn']), frozenset(), 'left')

def estado_objetivo(L, R):
    return len(L) == 0

def estado_seguros(L, R):
    if 'chicken' not in L and 'fox' in L and 'corn' in L:
        return False
    if 'chicken' not in R and 'fox' in R and 'corn' in R:
        return False
    return True

def acoes(estado):
    L, R, boat = estado
    proximas_acoes = []
    
    if boat == 'left':
        novos_itens = [
            (L - {'fox'}, R | {'fox'}, 'right'),
            (L - {'chicken'}, R | {'chicken'}, 'right'),
            (L - {'corn'}, R | {'corn'}, 'right'),
            (L, R, 'right')
        ]
    else:
        novos_itens = [
            (L | {'fox'}, R - {'fox'}, 'left'),
            (L | {'chicken'}, R - {'chicken'}, 'left'),
            (L | {'corn'}, R - {'corn'}, 'left'),
            (L, R, 'left')
        ]
    
    for novo_L, novo_R, novo_boat in novos_itens:
        if estado_seguros(novo_L, novo_R):
            proximas_acoes.append((novo_L, novo_R, novo_boat))
    
    return proximas_acoes

def busca_em_largura():
    inicio = estado_inicial()
    fila = deque([inicio])
    visitados = set()
    visitados.add(inicio)
    caminho = {inicio: None}
    
    while fila:
        estado_atual = fila.popleft()
        L, R, boat = estado_atual
        if estado_objetivo(L, R):
            return reconstrua_caminho(caminho, estado_atual)
        
        for proximo_estado in acoes(estado_atual):
            if proximo_estado not in visitados:
                visitados.add(proximo_estado)
                fila.append(proximo_estado)
                caminho[proximo_estado] = estado_atual
    
    return None

def reconstrua_caminho(caminho, estado_final):
    caminho_solucao = []
    while estado_final is not None:
        caminho_solucao.append(estado_final)
        estado_final = caminho[estado_final]
    return caminho_solucao[::-1]

solucao = busca_em_largura()
if solucao:
    print("Solução encontrada:")
    for passo in solucao:
        print(passo)
else:
    print("Nenhuma solução encontrada.")
```
